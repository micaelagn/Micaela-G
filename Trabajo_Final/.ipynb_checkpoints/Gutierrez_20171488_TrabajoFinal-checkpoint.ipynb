```python
#Se seleccionan las columnas de inter√©s
rfm1 = rfm[['Country name', 'year', 'Log GDP per capita',
            'Healthy life expectancy at birth', 'Perceptions of corruption']]
rfm1.head()
```

# <font color= darkcyan>An√°lisis</font>

<font color= slategray>Para este trabajo se utiliz√≥ la base de datos _'World Happiness Report'_. Para elaborar los gr√°ficos N¬∞ 1 y N¬∞ 2, se filtr√≥ la informaci√≥n para solo trabajar con la correspondiente a la del a√±o 2019. Con respecto al gr√°fico N¬∞ 3, este se enfoc√≥ en la variable de percepci√≥n de corrupci√≥n en los Pa√≠ses Bajos en un periodo de 15 a√±os. Se ten√≠a como objetivo visualizar si exist√≠a una correlaci√≥n entre los indicadores del PBI, percepci√≥n de corrupci√≥n y esperanza de vida. Asimismo, se plante√≥ representar c√≥mo la variable de percepci√≥n de corrupci√≥n pod√≠a variar a trav√©s del tiempo en un pa√≠s espec√≠fico. </font>

```python
#Gr√°fico 1: Correlaci√≥n entre Log GDP per c√°pita y esperanza de vida seg√∫n pa√≠ses en el a√±o 2019

alt.Chart(rfm2, height = 300, width = 500).mark_point(filled = True).encode(x= alt.X('Log GDP per capita',
title = 'Log PBI per c√°pita'), y = alt.Y('Healthy life expectancy at birth',
title = 'Esperanza de vida al nacer'), color = alt.Color('Healthy life expectancy at birth',
title = 'Esperanza de vida al nacer',scale=alt.Scale(scheme='goldgreen')), 
tooltip = ['Country name', 'Log GDP per capita', 'Healthy life expectancy at birth']).interactive(). properties(
title={
      "text": ["Gr√°fico N¬∞ 1: Correlaci√≥n entre Log GDP per c√°pita y esperanza de vida a nivel mundial en el a√±o 2019"] ,
    "subtitle": ["Fuente: Happiness Report"],
      "color": "midnightblue", "subtitleColor": "slategray"})
  
```

<font color= darkgreen>En el gr√°fico N¬∞ 1, es posible visualizar una correlaci√≥n positiva entre la variable Log GDP per c√°pita y la esperanza de vida, pues se observa que en los pa√≠ses donde hay un mayor PBI per c√°pita, la esperanza de vida al nacer tambi√©n es mayor.</font>

```python
#Gr√°fico 2: Correlaci√≥n entre Log GDP per c√°pita y percepci√≥n de corrupci√≥n seg√∫n pa√≠ses en el a√±o 2019

alt.Chart(rfm2, height = 400, width = 600).mark_point().encode(x= alt.X('Log GDP per capita',
title = 'Log PBI per c√°pita'), y = alt.Y('Perceptions of corruption',
title = 'Percepci√≥n de corrupci√≥n'), color = alt.Color('Perceptions of corruption',
title = 'Percepci√≥n de corrupci√≥n', scale=alt.Scale(scheme='purplebluegreen')),
tooltip = ['Country name', 'Perceptions of corruption']).interactive(). properties(
title={
      "text": ["Gr√°fico N¬∞ 2: Correlaci√≥n entre Log GDP per c√°pita y percepci√≥n de corrupci√≥n a nivel mundial en el a√±o 2019"] ,
      "subtitle": ["Fuente: Happiness Report"],
      "color": "midnightblue", "subtitleColor": "slategray"})
```

<font color= lightseagreen>En el gr√°fico N¬∞ 2 no es posible identificar una correlaci√≥n o tendencia clara entre las variables de Log PBI per c√°pita y la percepci√≥n de corrupci√≥n, pues se observa que es variada la percepci√≥n de corrupci√≥n en los pa√≠ses con mayor PBI per c√°pita, as√≠ como en aquellos con menor PBI.</font>

```python
g3a = alt.Chart(rfmpb).mark_line(color = 'thistle').encode(x = 'year', y = 'Perceptions of corruption').properties(
    width=700, height=200)
g3a
```

```python
g3b= alt.Chart(rfmpb).mark_point(color = 'rebeccapurple').encode(alt.X('year', title = 'A√±o'), y = alt.Y('Perceptions of corruption', title = 'Percepci√≥n de corrupci√≥n')).interactive()
g3b
```

```python
g3= g3a + g3b
```

```python
#Gr√°fico 3: Percepci√≥n de corrupci√≥n en Pa√≠ses Bajos durante el periodo 2005-2020

g3. properties(width=700, height=200,
title={
      "text": ["Gr√°fico N¬∞ 3: Percepci√≥n de corrupci√≥n en Pa√≠ses Bajos durante el periodo 2005-2020"] ,
      "subtitle": ["Fuente: Happiness Report"],
      "color": "midnightblue", "subtitleColor": "slategray"})
```

<font color= slateblue>En el gr√°fico N¬∞3, se observa c√≥mo ha cambiado el √≠ndice de percepci√≥n de corrupci√≥n en Pa√≠ses Bajos en un periodo de 16 a√±os (2004 - 2020). Es posible observar que con el paso de los a√±os este √≠ndice se ha ido reduciendo, es decir, los habitantes de los pa√≠ses bajos han percibido menores niveles de corrupci√≥n a medida que avanz√≥ el tiempo.</font>
